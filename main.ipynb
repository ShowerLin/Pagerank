```python
from pandas import DataFrame
import pandas as pd
import folium

station = pd.read_csv('201701-citibike-tripdata.csv')
station = station[0:726676]
```

```python
dlist = []
start_station = station.iloc[:,4].duplicated()
dlist = list(start_station)
l = []
for i in range(0,len(dlist)):
    if dlist[i]==True:
        l.append(i)
station = station.drop(l)
mapstation = DataFrame(station,columns = ['Start Station ID','Start Station Name','Start Station Latitude','Start Station Longitude'])
mapstation_pos = []

for i in range(0,station.index.size):
    mapstation_pos.append([mapstation.iloc[i,2],mapstation.iloc[i,3]])

import folium
m = folium.Map(location = mapstation_pos[1],zoom_start=13.5,tiles='Cartodb Positron')
for i in range(0,len(mapstation_pos)):
    folium.CircleMarker(location = mapstation_pos[i],radius = 1,color = '#FF8C00').add_to(m)
m
```

```python
PR_results = pd.read_csv("PR_results.csv")
PR_results = DataFrame(PR_results, columns = ['3#','Day3'])

tmp_pos = []

for i in range(1,604):
    a = PR_results.iloc[i,0]
    b = mapstation[mapstation['Start Station Name']==a]
    if len(b)==0:
        continue
    else:
        b.reset_index(drop = True,inplace=True)
        tmp_pos.append([b.iloc[0,2],b.iloc[0,3]])
    
m = folium.Map(location = tmp_pos[0],zoom_start=13.4,tiles='Cartodb Positron')
for i in range(0,len(tmp_pos)):
    folium.CircleMarker(location = tmp_pos[i],radius = 1500*PR_results.iloc[i,1],color = '#FF8C00').add_to(m)
m

```
